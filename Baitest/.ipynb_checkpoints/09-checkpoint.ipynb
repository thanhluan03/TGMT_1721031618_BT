```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_path = 'C:/Users/ASUS/Documents/file_anh/anh2.jpg'
# doc anh
im = Image.open(my_path)
plt.imshow(im)
```

```python
# chuyen sang anh xam
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
#tao mot hinh ve moi
plt.figure()
plt.hist(im_array.flatten(), bins=128)
plt.title('Bieu do Histogram của ảnh')
plt.xlabel('Giá trị của pixel')
plt.ylabel('Số lượng')
plt.show()
```

```python
#lam noi bat duong vien
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```

```python
# duong dan anh
my_path = 'C:/Users/ASUS/Documents/file_anh/anh3.jpg'

# doc anh va hien thi anh voi truc toa do 
im = Image.open(my_path).convert("L")

# hien thi anh tren truc toa do
plt.imshow(im)

# tao mot hinh ve moi
plt.figure()

# dat mau sac la xam
plt.gray()

# ve bieu do duong vien (contour) cua anh
plt.contour(im, origin="image")

# dam bao ty le truc x va y tren bieu do la bang nhau
plt.axis('equal')

# hiển thị ảnh
plt.show()
```
